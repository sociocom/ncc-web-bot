```python
def make_url_path(row):
    result = ""
    for i in range(1,6):
        url = row[f"URL{i}"]
        path = row[f"PATH{i}"]
        if pd.isnull(url) or url =="NaN" or url == "":
            result += ""
        elif pd.isnull(path) or path =="NaN" or path == "":
            result += "\t" + f"{str(url)}"
        else:
            result += "\t" + f"{str(path)}\n{str(url)}"
    return result.lstrip("\t").rstrip("\t")
```

```python
df_a["URL"] = df_a.apply(make_url_path, axis=1)
```
